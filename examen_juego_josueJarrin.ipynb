```python
# Ejercicio C) maquina de slots
# lets go gambling!!!!

import random
simbolos = []
opciones = ('@', '#', '$', '%','&')

class Jugador:
  premio = 0
  saldo = 0
  def __init__(self, nombre, monedas=100):
    self.nombre=nombre
    self.monedas=monedas

  def apostar(self):
    self.monedas= self.monedas-10
    for i in range(3):
      simbolo = random.choice(opciones)
      simbolos.append(simbolo)
    return simbolos

  def generarPremio(self):
    print(simbolos)
    for s in simbolos:
      if simbolos[0] == simbolos[1] or simbolos[0]==simbolos[2] or simbolos[1]==simbolos[2]:
        premio = 20
      elif simbolos[0]==simbolos[1] and simbolos[0]==simbolos[2] and simbolos[1]==simbolos[2]:
        premio = 50
      else:
        premio = 0

    if premio == 20:
      print("2 iguales, tu premio son 20, sigue apostando!!")
    elif  premio == 50:
      print("3 IGUALES!!! JACKPOT!!! 50 MONEDAS!")
    else:
      print("no tuviste simbolos iguales. premio es 0 monedas. sigue apostando.")
    return premio


  def recibirPremio(self, premio):
    saldo = self.monedas + premio
    return saldo

  def mostrarSaldo(self, saldo):
    print(saldo)

nombre = input("Escriba su nombre")
jugador1= Jugador(nombre, 100)
print(f"jugador {jugador1.nombre}")
print(f"monedas iniciales: {jugador1.monedas}")
jugar = input("quiere jugar si o no: ")
if jugar == "si":
  jugador1.apostar()
  jugador1.generarPremio()
  jugador1.recibirPremio(jugador1.premio)
  print(f"tus monedas despues de la partida son:")
  jugador1.mostrarSaldo(jugador1.saldo)
else:
  print("ya nada vuelva pronto")

```
